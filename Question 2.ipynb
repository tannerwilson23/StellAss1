```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (20, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
```

```python
plt.figure(figsize=(20,10))

labels = ['2GK','3.5GK','4GK','4.5GK','6GK' ]
dataA1 = np.loadtxt('foo_element_final.dat', skiprows=1, usecols = [0,1]);
plt.plot(dataA1[:,0],dataA1[:,1],label = labels[0], marker = '.')

dataA2 = np.loadtxt('a2element_final.dat', skiprows=1, usecols = [0,1]);
plt.plot(dataA2[:,0],dataA2[:,1],label = labels[1], marker = '.')

dataA3 = np.loadtxt('a3element_final.dat', skiprows=1, usecols = [0,1]);
plt.plot(dataA3[:,0],dataA3[:,1],label = labels[2], marker = '.')

dataA4 = np.loadtxt('a4element_final.dat', skiprows=1, usecols = [0,1]);
plt.plot(dataA4[:,0],dataA4[:,1],label = labels[3], marker = '.')

dataA5 = np.loadtxt('a5element_final.dat', skiprows=1, usecols = [0,1]);
plt.plot(dataA5[:,0],dataA5[:,1],label = labels[4], marker = '.')
plt.legend(loc = 'best')
plt.xticks(np.arange(min(dataA5[:,0]), max(dataA5[:,0])+1, 5.0))
plt.xlabel("Z")
plt.ylabel(r'X$_Z$')
```

```python
name = np.loadtxt('b1element_final.dat', skiprows=1, usecols= [2], dtype='str')

dataB1 = np.loadtxt('b1decayed.dat', usecols = [0,1, 2, 3]);

dataB2 = np.loadtxt('b2decayed.dat', usecols = [0,1, 2, 3]);

dataB3 = np.loadtxt('b3decayed.dat', usecols = [0,1, 2, 3]);

for a in [dataB1, dataB2, dataB3]:
    plt.figure(figsize=(20,10))
    for i in range(24,31):

        A = a[np.where(a[:,0]==i),1].flatten()
        Z = a[np.where(a[:,0]==i),0].flatten()
        decayed = a[np.where(a[:,0]==i),2].flatten()
        names  = name[i-1].capitalize()
        plt.plot(A, decayed, label = names, marker = '.')
    plt.ylim(1e-25,1)
    plt.legend(loc='best')
    plt.xlabel("A")
    plt.ylabel(r'X$_{^A Z}$')
    plt.yscale('log')
```

```python
name = np.loadtxt('c1element_final.dat', skiprows=1, usecols= [2], dtype='str')

dataC1 = np.loadtxt('c1decayed.dat', skiprows=1,usecols = [0,1, 2]);
dataC2 = np.loadtxt('c2decayed.dat', skiprows=1,usecols = [0,1, 2]);
dataC3 = np.loadtxt('c3decayed.dat', skiprows=1,usecols = [0,1, 2]);
dataB2 = np.loadtxt('b2decayed.dat', skiprows=1,usecols = [0,1, 2]);

for a in [dataC1,dataB2, dataC2, dataC3]:
    plt.figure(figsize = (20,10))
    plt.axvline(x=0.55, label = 'Z/A = 0.55', color = 'k')
    plt.axvline(x=0.46, label = 'Z/A = 0.46', color = 'k', linestyle = ':')
    plt.axvline(x=0.42, label = 'Z/A = 0.42', color = 'k', linestyle = '-.')
    plt.axvline(x=0.50, label = 'Z/A = 0.50', color = 'k', linestyle = '--')
    for i in range(10,50):
        A = a[np.where(a[:,0]==i),1].flatten()
        Z = a[np.where(a[:,0]==i),0].flatten()
        Nex = Z/A
        decayed = a[np.where(a[:,0]==i),2].flatten()
        names  = name[i-1].capitalize()
        plt.plot(Nex, decayed, label = names, marker = '.')
    plt.legend(loc='best')
    plt.xlabel("Z/A")
    plt.ylabel(r'X$_{^A Z}$')
    plt.yscale('log')
```
